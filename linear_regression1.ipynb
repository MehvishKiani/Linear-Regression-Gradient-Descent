```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5])  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable
```

```python
# Initialize parameters
theta0 = 0  # for the intercept
theta1 = 0  # for the slope
```

```python
# Hyperparameters
learning_rate = 0.01  # Learning rate
iterations = 1000   # Number of iterations
```

```python
# Function to calculate the cost (mean squared error)
def compute_cost(theta0, theta1, X, y):
    m = len(X)
    h = theta0 + theta1 * X
    J = (1 / (2 * m)) * np.sum((h - y) ** 2)
    return J
```

```python
# Gradient Descent
m = len(X)
for iteration in range(1000):
    # Compute predictions
    h = theta0 + theta1 * X
```

```python
# Lists to store the history of parameter updates and costs
theta0_history = [theta0]
theta1_history = [theta1]
cost_history = []
```

```python
# Gradient Descent
m = len(X)
for iteration in range(iterations):
    # Compute predictions
    h = theta0 + theta1 * X
```

```python
    # Compute gradients
    gradient0 = (1 / m) * np.sum(h - y)
    gradient1 = (1 / m) * np.sum((h - y) * X)

    # Update parameters using gradient descent
    theta0 = theta0 - learning_rate * gradient0
    theta1 = theta1 - learning_rate * gradient1
```

```python
# Calculate the cost
cost = compute_cost(theta0, theta1, X, y)

# Store parameter values and cost in history lists
theta0_history.append(theta0)
theta1_history.append(theta1)
cost_history.append(cost)
```

```python
# Final parameter values
print("Optimal parameters:")
print(f"theta0 = {theta0}")
print(f"theta1 = {theta1}")
```

```python
# Predict a new value
new_x = 5
predicted_y = theta0 + theta1 * new_x
print(f"Predicted value at x={new_x}: {predicted_y}")
```

```python
# Plot the data points
plt.scatter(X, y, label="Data")
```

```python
# Plot the best-fit line
plt.plot(X, theta0 + theta1 * X, color='red', label="Best Fit Line")
```

```python
# Plot the gradient descent path
plt.plot(theta0_history, theta1_history, 'bo-', label="Gradient Descent")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Gradient Descent")
plt.legend()
plt.show()
```
